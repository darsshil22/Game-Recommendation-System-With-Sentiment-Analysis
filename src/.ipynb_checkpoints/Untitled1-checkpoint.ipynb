# Step 1: Load and Inspect the Data
In this step, we will load the cleaned dataset and inspect the first few rows to understand its structure.



```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the cleaned dataset
df = pd.read_csv('D://AIDI//AIDI 1006//Assignment 4//data//video_game_reviews_with_sentiment.csv')

# Display the first few rows to check the data
df.head()
```

# Step 2: Map Sentiment Labels to Polarity Values
We will create a mapping of sentiment labels (`positive`, `neutral`, `negative`) to numeric values (1, 0, -1). This will allow us to convert the 'Sentiment Label' column into a 'Polarity' column with values corresponding to positive, neutral, or negative sentiment.


```python
# Mapping the 'Sentiment Label' to polarity values
sentiment_mapping = {
    'Positive': 1,
    'Neutral': 0,
    'Negative': -1
}

# Create a new column 'Polarity' based on the 'Sentiment Label' column
df['Polarity'] = df['Sentiment Label'].map(sentiment_mapping)

# Check if the 'Polarity' column was created correctly
df[['Sentiment Label', 'Polarity']].head()
```

# Step 3: Save the Updated Data
Now that we have created the 'Polarity' column, we will save the updated dataset with both sentiment labels and polarity values into a new CSV file for further use.


```python
# Save the updated dataset with the new 'Polarity' column
df.to_csv('D://AIDI//AIDI 1006//Assignment 4//data//video_game_reviews_with_sentiment_and_polarity.csv', index=False)

# Display confirmation message
print("\nSentiment labels converted to polarity values and saved as 'video_game_reviews_with_sentiment_and_polarity.csv'.")
```

# Plot the Distribution of Sentiment Labels
We will visualize the distribution of sentiment labels (positive, negative, and neutral).


```python
# Plot the distribution of sentiment labels
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Sentiment Label', palette='Set2')
plt.title('Distribution of Sentiment Labels')
plt.xlabel('Sentiment Label')
plt.ylabel('Count')
plt.show()
```

# Plot Sentiment vs User Rating
Let's visualize the relationship between sentiment labels and user ratings.


```python
# Create a boxplot to visualize the relationship between sentiment and user rating
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='Sentiment Label', y='User Rating', palette='Set1')
plt.title('Sentiment vs User Rating')
plt.xlabel('Sentiment Label')
plt.ylabel('User Rating')
plt.show()
```

```python
from wordcloud import WordCloud

# Combine all the reviews into a single string
all_reviews = ' '.join(df['User Review Text'].dropna())

# Generate a word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_reviews)

# Display the word cloud
plt.figure(figsize=(10, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for User Reviews')
plt.show()
```
